### Iterating

```python
import numpy as np
```

```python
a1= np.arange(10)
a2= np.arange(12).reshape(3, 4)
a3= np.arange(8).reshape(2, 2, 2)
```

```python
a1
```

```python
for i in a1:
  print(i)
```

```python
a2
```

```python
for i in a2:
  print(i)
```

```python
for i in a3:
  print(i)
```

```python
a4= np.arange(27).reshape(3, 3, 3)
a4
```

### Stacking

```python
#horizental stacking
a5= np.arange(16).reshape(4,4)
a6= np.arange(16, 32).reshape(4,4)
a6
```

```python
np.hstack((a5, a6))
```

```python
np.hstack((a6, a5, a6))
```

```python
np.hstack((a6, a5))
```

```python
#vertical stacking
np.vstack((a5, a6))
```

```python
np.vstack((a6, a5, a5))
```
