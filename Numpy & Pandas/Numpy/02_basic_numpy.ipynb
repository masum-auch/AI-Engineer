# NumPy Arrays Vs Python Lists (Sequences)

### Size change

* NumPy arrays are fixed-size when created.

* Python lists are dynamic, meaning they can be made larger or smaller later.

* To change the size of a NumPy array, a new array is created, and the old one is deleted.

### Data type

* All elements in a NumPy array must be of the same data type.

* Therefore, all elements in memory are of the same size.

* Python lists can hold mixed types.

### Mathematical and other operations

* Mathematical operations on large datasets can be performed quickly and easily using NumPy arrays.

* Doing the same thing with Python lists takes more code and time.

## Used in many packages

* Almost all Python packages, like SciPy, Pandas, Matplotlib, etc, use NumPy arrays.

* When given a Python list, they convert it to a NumPy array and then process it.

* The output is usually a NumPy array.

```python
# np.array
import numpy as np

a = np.array([1,2,3])
print(a)
```

```python
# 2D and 3D
b = np.array([[1,2,3],[4,5,6]])
print(b)

c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
```

```python
# dtype try
np.array([7,1,3],dtype=bool)
np.array([1,2,3],dtype=float)
```

```python
# np.arange
np.arange(1,11).reshape(5, 2)
```

```python
np.arange(25).reshape(5,5)
```

```python
# with reshape
np.arange(16).reshape(2, 2, 2, 2)
```
