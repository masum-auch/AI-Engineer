# Array Functions

```python
a6 = np.random.random((3,3))
a7 = np.round(a6*100)
a7
```

```python
# min/max/sum/prod
# 0 > col or 1 > row

np.max(a7, axis= 0)
```

```python
np.prod(a7, axis= 0)
```

```python
# mean/median/std/var

# np.mean(a7)
np.median(a7)
# np.std(a7)
# np.var(a7)
```

```python
# trigonomoetric functions
np.sin(a6)
```

```python
# dot product
a8 = np.arange(12).reshape(3,4)
a9 = np.arange(12,24).reshape(4,3)

np.dot(a8,a9)
```

```python
# log and exponents
np.exp(a8)
```

```python
# round/floor/ceil

np.floor(np.random.random((2,3))*100)
```
