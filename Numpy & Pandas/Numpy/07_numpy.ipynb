## Advanced Indexing

```python
#import numpy lab
import numpy as np
```

```python
# Boolean Indexing
a = np.random.randint(1,100,24).reshape(6,4)
a
```

```python
# find all numbers greater than 50
a[a > 50]
```

```python
# find out even numbers
a[a % 2 == 0]
```

```python
# find all numbers greater than 50 and are even

a[(a > 50) & (a % 2 == 0)]
```

```python
# find all numbers not divisible by 7
a[~(a % 7 == 0)]
```
