```python
mixed_df= df.copy()
```

```python
mixed_df.head()
```

```python
mixed_df['Cabin_Num']= mixed_df['Cabin'].str.extract('(\d+)')
mixed_df['Cabin_Cate']= mixed_df['Cabin'].str[0]
```

```python
mixed_df.sample(5)
```

```python
mixed_df['Embarked']= mixed_df['Embarked'].fillna(mixed_df['Embarked'].mode()[0])
```

```python
mixed_df.isna().sum()
```

```python
mixed_df.drop(columns= 'Cabin', inplace= True)
```

```python
mixed_df['Cabin_Num']= mixed_df['Cabin_Num'].fillna(-1)
mixed_df['Cabin_Cate']= mixed_df['Cabin_Cate'].fillna('Missing')
```

```python
mixed_df.head()
```

```python
mixed_df['Ticket_Num']= mixed_df['Ticket'].apply(lambda s: s.split()[-1])
mixed_df['Ticket_Num']= pd.to_numeric(mixed_df['Ticket_Num'], errors= 'coerce', downcast= 'integer')


mixed_df['Ticket_Cate']= mixed_df['Ticket'].apply(lambda s: s.split()[0])
mixed_df['Ticket_Cate']= np.where(mixed_df['Ticket_Cate'].str.isdigit(), np.nan, mixed_df['Ticket_Cate'])
```

```python
mixed_df.drop(columns= 'Ticket', inplace= True)
```
