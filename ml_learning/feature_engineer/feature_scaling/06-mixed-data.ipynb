```python
#fill Age, Embarked missing Data

df['Age']= df['Age'].fillna(df['Age'].median(), inplace= True)
df['Embarked']= df['Embarked'].fillna(df['Embarked'].mode()[0], inplace= True)
```

```python
#Handle mixed Data | Cabin| the mixed columns part 2 cabin numeric and categorical

df['Cabin_Num']= df['Cabin'].str.extract('(\d+)')
df['Cabin_Cate']= df['Cabin'].str[0]

#the original Cabin columns no Unnecessary, so drop the columns

df.drop(columns= 'Cabin', inplace= True)
```

```python
df.sample(5)
```

```python
df.isna().sum()[df.isna().sum() > 0]
```

```python
# after creating columns Cabin_Num and Cabin_Cate then i see missing value. So handle the missing values

df['Cabin_Num']= df['Cabin_Num'].fillna(-1)
df['Cabin_Cate']= df['Cabin_Cate'].fillna('Missing')
```

```python
#mixed data against ticket

# df['Ticket'].unique()

#create numeric columns
df['ticket_num']= df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num']= pd.to_numeric(df['ticket_num'], errors= 'coerce', downcast= 'integer')

#create the categorical columns
df['ticket_cate']= df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cate']= np.where(df['ticket_cate'].str.isdigit(), np.nan, df['ticket_cate'])

#the remove original columns

df.drop(columns= 'Ticket', inplace= True)

# after creating column ticket_num, I see a missing value. So handle the missing values

df['ticket_cate']= df['ticket_cate'].fillna('Missing')
```

```python
# # ensure string + uppercase
# df['ticket_cate'] = (df['ticket_cate'].astype(str).str.upper().str.strip())

# # keep slash/dot as separator, remove extra spaces only
# df['ticket_cate'] = df['ticket_cate'].str.replace(r'\s+', '', regex=True)

# # extract ONLY alphabet prefix (number drop)
# df['ticket_prefix'] = df['ticket_cate'].str.extract(r'^([A-Z]+)')


df['ticket_prefix'] = df['ticket_cate'].str.extract(r'^([A-Z./]+)')
df['ticket_prefix'] = df['ticket_prefix'].fillna('NUM')
```

```python
freq = df['ticket_prefix'].value_counts()
rare = freq[freq < 10].index

df['ticket_prefix'] = df['ticket_prefix'].replace(rare, 'Other')
```
