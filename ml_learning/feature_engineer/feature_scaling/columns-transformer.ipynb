# Learn with columns transformer.

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('/kaggle/input/covid-toy/covid_toy.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns= ['has_covid']), df['has_covid'], test_size= 0.2)
```

```python
X_train.shape
```

# Using SimpleImputer

```python
si= SimpleImputer()

X_train_fever= si.fit_transform(X_train[['fever']])
X_test_fever= si.fit_transform(X_test[['fever']])

X_train_fever.shape
```

```python
X_test_fever.shape
```

# Using Ordinal Encoder

```python
oe = OrdinalEncoder(categories= [['Mild', 'Strong']])

X_train_cough= oe.fit_transform(X_train[['cough']])
#test
X_test_cough= oe.fit_transform(X_test[['cough']])

X_train_cough.shape
```

```python
# X_train_cough
```

# Using One Hot Encoding

```python
ohe= OneHotEncoder(drop= 'first', sparse= False, dtype= int)

X_train_new= ohe.fit_transform(X_train[['gender', 'city']])
#test
X_test_new= ohe.fit_transform(X_test[['gender', 'city']])

X_train_new.shape
```

```python
# X_train_new
```

```python
#extracting age
X_train_age= X_train.drop(columns= ['gender', 'fever', 'cough', 'city']).values

X_test_age= X_test.drop(columns= ['gender', 'fever', 'cough', 'city']).values

X_train_age.shape
```

```python
X_train_transformer= np.concatenate((X_train_age, X_train_fever, X_train_cough, X_train_new), axis= 1)

X_test_transformer= np.concatenate((X_train_age, X_train_fever, X_train_cough, X_train_new), axis= 1)

X_train_transformer.shape
```

# Using columns Transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer= ColumnTransformer(transformers= [
    ('tnf1', SimpleImputer(), ['fever']),
    ('tnf2', OrdinalEncoder(categories= [['Mild', 'Strong']]), ['cough']),
    ('tnf3', OneHotEncoder(drop= 'first', sparse= False, dtype= int), ['gender', 'city'])
    
], remainder= 'passthrough')
```

```python
transformer.fit_transform(X_train).shape
```
