# warking with date

```python
date['date']= pd.to_datetime(date['date'])
```

```python
date['date_year']= date['date'].dt.year

date.sample(5)
```

```python
date['date_month']= date['date'].dt.month
date.sample(5)
```

```python
date['date_month_name']= date['date'].dt.month_name()
date.sample(5)
```

```python
date['date_inday']= date['date'].dt.day
date.sample(5)
```

```python
date['date_dow']= date['date'].dt.dayofweek
date.sample(5)
```

```python
date['date_inday_name']= date['date'].dt.day_name()
date.sample(5)
```

```python
date.drop(columns= ['product_id', 'city_id', 'orders'], inplace= True)
```

```python
date.head(10)
```

```python
date['date_is_weekend']= np.where(date['date_inday_name'].isin(['Sunday', 'Saturday']), 1, 0)
date.head(10)
```

# Extract week of the year

```python
date['date_week']= date['date'].dt.isocalendar().week
date.head(4)
```

```python
date['quarter']= date['date'].dt.quarter
date.head(5)
```

```python
date['semester']= np.where(date['quarter'].isin([1, 2]), 1, 2)
date.head(5)
```
