# This Part is Handling Missing Data | KNN Imputer

```python
from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/kaggle/input/train-data/train.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.isna().mean() * 100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.sample(5)
```

```python
knn= KNNImputer(n_neighbors= 3)

X_train_k= knn.fit_transform(X_train)

X_test_k= knn.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_k,y_train)

y_pred = lr.predict(X_test_k)

accuracy_score(y_test,y_pred)
```

```python
# Comparision with Simple Imputer --> mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2,y_train)

y_pred2 = lr.predict(X_test_trf2)

accuracy_score(y_test,y_pred2)
```
