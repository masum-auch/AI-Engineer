None of the patients were previously using NovoDra (a popular injectable insulin) or Auralin (an investigational oral insulin) as their primary source of insulin. All were experiencing increasing HbA1c levels.

All 350 patients were treated with Novodra to establish a baseline HbA1c level and insulin dose. After 4 weeks, which isn’t enough time to capture all the changes in HbA1c that the switch can attribute to Auralin or Novodra:

- 175 patients switched to Auralin for 24 weeks
- 175 patients continued using Novodra for 24 weeks
