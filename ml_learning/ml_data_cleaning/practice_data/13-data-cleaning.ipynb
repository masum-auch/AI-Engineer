## 4. Data Cleaning 

```python
#standardize column name
df.columns= df.columns.str.lower().str.replace(" ", "_")

#drop the duplicated columns
df.drop_duplicates(inplace= True)

#fix missing value
def safe_fill(df, cols, value= 'unknown', numeric= False):
    for col in cols:
        if df[col].isna().sum()> 0:
            if numeric:
                df[col]= df[col].fillna(df[col].median())
            else:
                df[col]= df[col].fillna(value)
    return df

#fix the categorical value space and /
df['category']= df['category'].str.lower().str.strip().str.replace('/', ' ', regex=True)
df['primary_skills']= df['primary_skills'].str.lower().str.strip().str.replace('/', ',', regex=True)

#clean the annual_income_usd
df['annual_income_usd']= pd.to_numeric(
    df['annual_income_usd'].str.replace(r'[$,]', '', regex= True), 
    errors= 'coerce'
)
```
