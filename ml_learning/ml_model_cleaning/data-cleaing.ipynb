```python
house_data['total_bedrooms']= house_data['total_bedrooms'].fillna(house_data['total_bedrooms'].mean())
```

```python
house_data['ocean_proximity'].value_counts()
```

```python
cal_col= ['ocean_proximity']
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe= OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded= ohe.fit_transform(house_data[cal_col])
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns=ohe.get_feature_names_out(cal_col)
)
```

```python
house_data = house_data.drop(columns=cal_col).reset_index(drop=True)
house_data = pd.concat([house_data, encoded_df], axis=1)
```
